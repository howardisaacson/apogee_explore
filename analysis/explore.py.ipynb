```python
import astropy.io.fits as fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
f="apStar-r8-2M18521876+4525392.fits"
hdu=fits.open(f)
```

```python
hdu.info()
```

```python
head = hdu[0].header
flux = hdu[6].data
wav  = hdu[2].data # not wavelength
w0 = head['CRVAL1']
w_disp = head['CDELT1']
w_disp
npix = 8575
wls = np.arange(w0,w0+w_disp*npix,w_disp)
wls_lin = 10**wls
```

```python
len(flux)
#head
```

```python
plt.figure(figsize=[12,8])
plt.plot(wls_lin,flux[0,:])
plt.xlabel('Angstrom')
plt.ylabel('Flux Calibrated [ 10^-17 ergs/s/cm^2/Ang]')
plt.show()
```
